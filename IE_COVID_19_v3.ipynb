# Looking at the outbreak

There are months of data, but we only care about when the number of cases started to grow. We define outbreak as whenever the number of cases exceeded certain threshold and for data quality we remove data without Region associated. In this case, we are using 5 Deaths.

## Grouping all data by country (not regions)

```python
data_bkp = data.copy()
```

---

```python
data = data_bkp.copy()
```

```python
data = data[~(data['RegionCode'].isna())]
```

```python
data = data[['CountryName','Date','Confirmed','Deaths']].groupby(['CountryName','Date']) \
                             .sum().sort_values(by = ['Date'], ascending = False)
data
```

```python
def get_outbreak_mask(data: pd.DataFrame, threshold: int = 10, concept: str = 'Confirmed'):
    ''' Returns a mask for > N confirmed cases '''
    return (data[concept] > threshold) 

cols = ['CountryCode', 'CountryName', 'Confirmed', 'Deaths']
# Get data only after the outbreak begun
data = data[get_outbreak_mask(data,threshold=5,concept='Deaths')]

data.head()
```

```python
data = data.sort_values(['CountryName','Date'], ascending = (True, True))
data.head()
```

```python
data.to_csv("cov19.csv")
```

### Some moving average variables


```python
df = df.sort_values(['CountryName','day'], ascending = (True, True))
```

```python
# Group df by df.platoon, then apply a rolling mean lambda function to df.casualties
df['DailyDeaths_MA_3'] = df.groupby('CountryName')['DailyDeaths'].apply(lambda x:x.rolling(center=False,window=3).mean())
df['DailyDeaths_MA_7'] = df.groupby('CountryName')['DailyDeaths'].apply(lambda x:x.rolling(center=False,window=7).mean())
df['DailyDeaths_MA_14'] = df.groupby('CountryName')['DailyDeaths'].apply(lambda x:x.rolling(center=False,window=14).mean())
df['DailyDeaths_MA_21'] = df.groupby('CountryName')['DailyDeaths'].apply(lambda x:x.rolling(center=False,window=21).mean())
```

```python
df[['day','CountryName','DailyDeaths_MA_3','DailyDeaths_MA_7','DailyDeaths_MA_14','DailyDeaths_MA_21']].head()
```
